```python
from ModelKK import *
from SR_utils import *
import numpy as np
from keras.models import load_model
import h5py
import matplotlib.pyplot as plt
```

```python
VModel = VoiceModel()
VModel.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer='adam')
```

```python
utterance_length = 35
train_batch = get_mfcc_batch('../data/', 250, utterance_length)

for iterations in range(30):
    
    X, Y = next(train_batch)
    X = np.asarray(X)
    Y = np.asarray(Y)
    VModel.fit(X, Y, epochs=50, batch_size=250)
```

```python
VModel.save('VModel.h5')
```

```python
P = extract_mfcc('../data/6_nicolas_45.wav', utterance_length)
P = np.expand_dims(P, axis=0)
prediction_digit = VModel.predict(P)
print("Digit predicted: ", np.argmax(prediction_digit))
```
